```python
import pandas as pd

df_store = pd.read_csv("avocadostore.csv")
```

```python
df_store.head(4)
```

```python
# change column names to all lower cases
df_store.rename(columns=str.lower, inplace=True)
df_store.head(2)
```

```python
# Option 1: remove all blank spaces in column names
df_store.rename(columns={col: col.replace(' ', '') for col in df_store.columns}, inplace=True)
df_store.head(2)
```

```python
# Option 2 replace blank space with _ in column names
df_store.rename(columns={col: col.replace(' ', '_') for col in df_store.columns}, inplace=True)
df_store.head(2)
```
